```python
import re
def extract_value_unit(unit_of_measure):
    if pd.isna(unit_of_measure):
        return None, None

    match = re.match(r"(\d+(\.\d+)?)\s*([a-zA-Z/]+)", unit_of_measure)
    if match:
        value = float(match.group(1))
        unit = match.group(3)
        return value, unit
    return None, None
```

```python
# Apply the function to the unitOfMeasure column
df_cleaned[['value', 'unit']] = df_cleaned['unitOfMeasure'].apply(lambda x: pd.Series(extract_value_unit(x)))
```

```python
# Convert the 'unit' column to a categorical type
df_cleaned['unit'] = df_cleaned['unit'].astype('category')
```

```python
# Apply label encoding to the 'unit' column
df_cleaned['unit_numeric'] = df_cleaned['unit'].cat.codes
```
